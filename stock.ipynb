```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time

def fetch_news_with_retries(ticker, max_retries=5, backoff_factor=1):
    for attempt in range(max_retries):
        try:
            url = f"https://news.google.com/search?q={ticker}"
            response = requests.get(url, timeout=10)
            response.raise_for_status()
            soup = BeautifulSoup(response.content, "html.parser")
            headlines = [h.text for h in soup.find_all('a', class_='DY5T1d')]
            return headlines[:10]
        except requests.exceptions.RequestException as e:
            print(f"Error fetching news for {ticker}: {e}. Retrying...")
            time.sleep(backoff_factor * (2 ** attempt))
    print(f"Failed to fetch news for {ticker} after {max_retries} attempts.")
    return []
news_data = {ticker: fetch_news_with_retries(ticker) for ticker in tickers}
print(news_data)
```
